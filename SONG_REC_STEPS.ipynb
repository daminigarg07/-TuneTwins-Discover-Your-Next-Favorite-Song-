```python
import pandas as pd
import requests
import spotipy.util as util
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
from sklearn.metrics.pairwise import cosine_similarity
import warnings
warnings.filterwarnings('ignore')
```

```python
# Get your client id, secret key, and redirect URI from your application https://developer.spotify.com/dashboard/applications
client_id ='CLIENT ID' #<----------------- YOUR ID HERE
    
client_secret = "CLIENT SECRET" #<----------------- YOUR SECRET HERE
username = 'USERNAME' #<----------------- YOUR USERNAME HERE
scope = 'SCOPE' #<------------- depends on your needs, you can select different scope https://developer.spotify.com/documentation/general/guides/scopes/
redirect_uri = 'REDIRECT URL'#<----------------- PUT REDIRECT URI HERE
```

```python
from spotipy.oauth2 import SpotifyOAuth
```

```python
# Authenticate and initialize Spotify client
sp_oauth = SpotifyOAuth(client_id=client_id,
                        client_secret=client_secret,
                        redirect_uri=redirect_uri,
                        scope=scope)

# Get the access token (it returns a string directly if as_dict=False)
token_info = sp_oauth.get_access_token(as_dict=True)  # default is as_dict=True

# Initialize Spotify client with access token
sp = spotipy.Spotify(auth=token_info['access_token'])
```

```python
artist=input("artist you like:")
```

```python
# Search for the artist
artist_info = sp.search(artist, limit=1, offset=0, type='artist', market=None)
if not artist_info['artists']['items']:
    print("Artist not found.")
else:
    print(artist_info)
```

```python
artist_info = artist_info['artists']['items'][0]
artist_id = artist_info['id']
artist_id
```

```python
# Get albums by the artist
albums = sp.artist_albums(artist_id, album_type=["album", "single", "appears_on"], limit=50)
albums_df = pd.DataFrame(albums['items'])
albums_df.head()
```

```python
tracks_df = pd.DataFrame()
    
if not albums_df.empty:
    for album_id in albums_df['id']:
        album_tracks = sp.album_tracks(album_id, limit=50, offset=0, market=None)
        tracks_df = pd.concat([tracks_df, pd.DataFrame(album_tracks['items'])])
else:
    print("Nothing found")
        
tracks_df = tracks_df.reset_index(drop=True)  
tracks_df.head()
```

```python
# Get audio features for the tracks
track_ids = tracks_df['id']
chunk_size = 50
chunks = [track_ids[i:i + chunk_size] for i in range(0, len(track_ids), chunk_size)]

# Initialize a list to store the audio features
audio_features = []

# Retrieve audio features for each chunk of track IDs
for chunk in chunks:
    audio_features_chunk = sp.audio_features(chunk)
    audio_features.extend(audio_features_chunk)
audio_features_df = pd.DataFrame(audio_features)  
audio_features_df.head()
```

```python
# Combine track data with audio features
merged_df = tracks_df.merge(audio_features_df, left_on='id', right_on='id', suffixes=('', '_audio'))
merged_df.head()
```

```python
 # Perform recommendation logic (example: find similar tracks)
recommendation_df = merged_df[['name','id', 'danceability', 'energy', 'key', 'loudness', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']]
recommendation_df.head()
```

```python
cat =sp.categories(country=None, locale=None, limit=50, offset=0)
cat_df=pd.DataFrame(cat['categories']['items'])
cat_df.head()
```

```python
def rec(song):
    song = song.lower()
    recommendation_df['name'] = recommendation_df['name'].str.lower()
    x = recommendation_df.drop('name',axis=1)
    x = x.drop('id',axis=1)
    matrix = x
    similarity = cosine_similarity(matrix)
    print("Looking for song like:", song)
    
    matching_rows = recommendation_df[recommendation_df['name'].str.contains(song)]
    
    if matching_rows.empty:
        print("Song not found.")
        return
    
    idx = matching_rows.index[0]
    distances = sorted(list(enumerate(similarity[idx])), reverse=True, key=lambda x:x[1]) 
    
    songs = []
    for m_id in distances[1:11]:
        songs.append(recommendation_df.loc[m_id[0],'name'])
    recommendations=[]
    for x in songs:
        # check if exists in unique_list or not
        if x not in recommendations:
            recommendations.append(x)
    
    return recommendations
```

```python
song=input("Enter a song you like:")
rec(song)
```

```python
matching_rows = recommendation_df[recommendation_df['name'].str.contains(song)]
track_id = str(matching_rows.iloc[0]['id'])
track_id
```
